```python
import subprocess
import platform
import ipaddress
```

```python
network = ipaddress.ip_network('10.0.0.0/24', strict=False)
```

```python
alive_hosts = []

for ip in network.hosts():
    ip_str = str(ip)
    # Commande ping selon OS
    if platform.system().lower() == 'windows':
        cmd = ['ping', '-n', '1', '-w', '100', ip_str]
    else:
        cmd = ['ping', '-c', '1', '-W', '1', ip_str]
    try:
        result = subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        if result.returncode == 0:
            alive_hosts.append(ip_str)
    except Exception:
        pass

print('Machines actives sur le réseau:')
for host in alive_hosts:
    print(host)
```

```python
import subprocess
import platform

def get_wifi_interface_linux():
    try:
        result = subprocess.run(['iw', 'dev'], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
        for line in result.stdout.splitlines():
            if line.strip().startswith('Interface'):
                return line.strip().split()[1]
    except Exception:
        pass
    return 'wlan0'  # Valeur par défaut

system = platform.system().lower()
if system == 'darwin':
    print("ATTENTION : La commande 'airport' est dépréciée sur macOS. Pour des diagnostics avancés, utilisez 'wdutil diagnose'.")
    try:
        result = subprocess.run([
            '/System/Library/PrivateFrameworks/Apple80211.framework/Versions/Current/Resources/airport', '-s'
        ], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
        if result.returncode == 0:
            print('Réseaux WiFi disponibles:')
            print(result.stdout)
        else:
            print('Erreur lors de la récupération des réseaux WiFi:')
            print(result.stderr)
    except Exception as e:
        print(f'Erreur: {e}')
elif system == 'linux':
    interface = get_wifi_interface_linux()
    print(f"Scan des réseaux WiFi sur l'interface {interface}...")
    try:
        result = subprocess.run(['sudo', 'iwlist', interface, 'scan'], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
        if result.returncode == 0:
            print('Réseaux WiFi disponibles:')
            for line in result.stdout.splitlines():
                if 'ESSID' in line:
                    print(line.strip())
        else:
            print('Erreur lors de la récupération des réseaux WiFi:')
            print(result.stderr)
    except Exception as e:
        print(f'Erreur: {e}')
else:
    print('Système non supporté pour le scan WiFi automatique.')
```
